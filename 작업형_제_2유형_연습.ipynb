```python
import pandas as pd
import numpy as np
```

```python
X_test=pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_x_test.csv')
X_train=pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_x_train.csv')
y_train=pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_y_train.csv')
```

```python
COL_DEL=['ID']
COL_NUM=['Age','AnnualIncome','FamilyMembers']
COL_CAT=['Employment Type','GraduateOrNot','ChronicDiseases','FrequentFlyer','EverTravelledAbroad']
COL_Y=['TravelInsurance']
```

```python
X_train[COL_NUM].describe()
```

```python
X_train[COL_CAT]=X_train[COL_CAT].astype(str)
X_test[COL_CAT]=X_test[COL_CAT].astype(str)
```

```python
X_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
X=pd.concat([X_train,X_test])
for i in COL_CAT:
  le = LabelEncoder()
  le.fit(X[i])
  X_train[i]=le.transform(X_train[i])
  X_test[i]=le.transform(X_test[i])
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val = train_test_split(X_train[COL_NUM+COL_CAT],y_train[COL_Y],test_size=0.3,stratify=y_train[COL_Y])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_tr[COL_NUM]=scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM]=scaler.transform(X_val[COL_NUM])
X_test[COL_NUM]=scaler.transform(X_test[COL_NUM])
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_tr,y_tr)
```

```python
y_pred = model.predict_proba(X_val)
y_pred
```

```python
from sklearn.metrics import roc_auc_score
```

```python
score = roc_auc_score(y_val[COL_Y],y_pred[:,1])
score
```

```python
pred = model.predict_proba(X_test[COL_NUM+COL_CAT])
pred
```
