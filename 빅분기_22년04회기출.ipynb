#작업형 제 2유형
고객이 속한 세그먼트를 예측하고 예측 결과를 csv 파일로 제출하시오.<br>
예측값의 macro_f1 결과를 통해 최종 점수 반영 예정

```python
X_test = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/404_x_test.csv')
X_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/404_x_train.csv')
y_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/404_y_train.csv')
```

```python
X_train['Spending_Score'].value_counts(dropna=False)
```

```python
X_train['Profession'].value_counts(dropna=False)
```

```python
COL_DEL=['ID']
COL_NUM=['Age','Work_Experience','Family_Size']
COL_CAT=['Gender','Ever_Married','Graduated','Profession','Spending_Score']
COL_Y=['Segmentation']
```

```python
from sklearn.preprocessing import LabelEncoder
X=pd.concat([X_train,X_test])
for i in COL_CAT:
  le = LabelEncoder()
  le.fit(X[i])
  X_train[i]=le.transform(X_train[i])
  X_test[i]=le.transform(X_test[i])
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val = train_test_split(X_train[COL_NUM+COL_CAT],y_train[COL_Y],test_size=0.3,stratify=y_train[COL_Y])
```

```python
X_tr
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_tr[COL_NUM]=scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM]=scaler.transform(X_val[COL_NUM])
X_test[COL_NUM]=scaler.transform(X_test[COL_NUM])
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier()
model_rfc.fit(X_tr,y_tr)
```

```python
y_pred_rfc=model_rfc.predict(X_val)
```

```python
#f1-score
from sklearn.metrics import f1_score

score_rfc = f1_score(y_val,y_pred_rfc,average='macro')
print(score_rfc)
```

```python
pred=model_rfc.predict(X_test[COL_NUM+COL_CAT])
result=pd.DataFrame({'ID':X_test.ID,'Segementation':pred})
result
```
