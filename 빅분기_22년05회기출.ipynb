##1-2
주어진 BMI 데이터 세트에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력하시오<br>

bmi : 몸무게(kg)%키(m) => 몸무게(kg)/키(m)^2<br>
18.5 이상 23 미만 : 정상<br>
23 이상 25 미만 : 과체중

```python
data = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/502_bmi.csv')
data
```

```python
data.info()
```

```python
data['bmi']=data['Weight']/(data['Height']/100)**2
data
```

```python
data['bmi_group']=np.where(data['bmi']<18.5,'저체중',
                  np.where(data['bmi']<23,'정상',
                  np.where(data['bmi']<25,'과체중',
                  np.where(data['bmi']<30,'경도비만',
                  np.where(data['bmi']>=30,'중증도비만',np.nan)))))
data
```

```python
data['bmi_group'].value_counts(dropna=False)
```

```python
int(len(data[data['bmi_group']=='정상'])-len(data[data['bmi_group']=='과체중']))
```

#작업형 제 2유형
중고차의 판매 가격을 예측하고 해당 예측 결과를 csv 파일로 제출하시오.

```python
X_test = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/504_x_test.csv')
X_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/504_x_train.csv')
y_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/504_y_train.csv')
```

```python
X_train.describe()
```

```python
X_train['year']=2023-X_train['year']
X_test['year']=2023-X_test['year']
```

```python
X_train
```

```python
X_test
```

```python
X_train.describe()
```

```python
COL_DEL=['id']
COL_NUM=['year','mileage','tax',	'mpg','engineSize']
COL_CAT=['model','transmission','fuelType']
COL_Y=['price']
```

```python
X_train.drop(COL_DEL,axis=1,inplace=True)
X_test.drop(COL_DEL,axis=1,inplace=True)
```

```python
X_train
```

```python
X_train.describe()
```

```python
def outlier(x):
  q1,q3 = x.quantile([0.25,0.75])
  iqr= q3-q1
  upper = q3+1.5*iqr
  lower = q1-1.5*iqr
  return str(len(x[(x<lower)|(x>upper)]))
X_train[COL_NUM].apply(outlier)
```

```python
X_train['mileage'].quantile([q/100 for q in range(0,26)])
```

```python
X_train['mileage']=np.where(X_train['mileage']<1000,1000,X_train['mileage'])
```

```python
X_train['tax'].quantile([q/100 for q in range(0,26)])
```

```python
X_train['mpg'].quantile([q/100 for q in range(90,101)])
```

```python
X_train['mpg']=np.where(X_train['mpg']>100,100,X_train['mpg'])
```

```python
X_train.describe()
```

```python
X_train.info()
```

```python
# COL_NUM=['year','mileage','tax',	'mpg','engineSize']
# COL_CAT=['model','transmission','fuelType']
# COL_Y=['price']
from sklearn.preprocessing import LabelEncoder

X=pd.concat([X_train,X_test])
for i in COL_CAT:
  le = LabelEncoder()
  le.fit(X[i])
  X_train[i]=le.transform(X_train[i])
  X_test[i]=le.transform(X_test[i])
```

```python
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val = train_test_split(X_train,y_train,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_tr[COL_NUM] = scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.transform(X_test[COL_NUM])
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(X_tr,y_tr['price'])
```

```python
X_tr
```

```python
y_tr
```

```python
X_val
```

```python
from sklearn.metrics import mean_squared_error
y_pred_rf = model_rf.predict(X_val)
y_pred_rf
```

```python
rmse = mean_squared_error(y_val['price'],y_pred_rf,squared=False)
rmse
```
