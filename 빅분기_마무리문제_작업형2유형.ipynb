```python
import pandas as pd
import numpy as np

X_test = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/stellar_X_test.csv')
X_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/stellar_X_train.csv')
y_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/stellar_y_train.csv')
```

```python
def outlier(x):
  q1,q3 = x.quantile([0.25,0.75])
  iqr=q3-q1
  low = q1-1.5*iqr
  high = q3+1.5*iqr
  return str(len(x[(x<low)|(x>high)]))
```

```python
X_train.apply(outlier)
```

```python
y_train
```

```python
# -9999 이상치 제거, X 셋에서 제거된 행은 y셋도 제거되어야 함

data = pd.concat([X_train,y_train],axis=1)
data = data[data['u']!=-9999]
data
```

```python
data.describe()
```

```python
X_train = data.drop(['galaxy'],axis=1)
y_train=data['galaxy']
```

```python
X_train.reset_index(drop=True)
y_train.reset_index(drop=True)
```

```python
X_train.describe()
```

```python
X_train['redshift'].quantile([q/100 for q in range(90,101)])
```

```python
#3이 넘는 2프로 파생 변수 추가하여 범주형 변수로 변환 -> X dataset에 모두 추가
X_train['redshift_upper3'] = np.where(X_train['redshift']>=3,1,0)
X_test['redshift_upper3'] = np.where(X_test['redshift']>=3,1,0)
```

```python
X_train
```

```python
COL_DEL =[]
COL_NUM = ['alpha','delta','u','g','r','i','z','redshift']
COL_CAT =['run_ID',	'rerun_ID',	'cam_col',	'field_ID',	'plate', 'redshift_upper3']
COL_Y = ['galaxy']
```

```python
X_train[COL_CAT]=X_train[COL_CAT].astype(str) #범주형 변수 타입 변환
X_test[COL_CAT]=X_test[COL_CAT].astype(str)
```

```python
train_df = pd.concat([X_train,y_train],axis=1)

for i in COL_NUM:
  print("-"*50)
  print('column name :  ',i)
  print(train_df.groupby('galaxy')[i].describe(),end='\n\n')
```

수치형 변수 중 의미 있어 보이는 변수 = redshift<br>
다른 변수는 통계량이 galaxy 여부에 상관없이 비슷함

```python
for i in COL_CAT:
  print("-"*50)
  print('column name : ',i)
  print(train_df.groupby(i)['galaxy'].mean())
```

cam_col,redshift_upper3 이외의 다른 변수의 경우 범주가 너무 많아 사용 x

```python
from sklearn.preprocessing import LabelEncoder

X= pd.concat([X_train,X_test]) #위 아래 병합
```

```python
for i in COL_CAT:
  le = LabelEncoder()
  le.fit(X[i]) # items=['d','s','a','e'] le.fit(items)
  X_train[i] = le.transform(X_train[i])
  X_test[i] = le.transform(X_test[i])
```

```python
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val = train_test_split(X_train,y_train,test_size=0.3,stratify=y_train) #범주형 변수 클래스 분포 맞추기 stratify
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_tr[COL_NUM] = scaler.fit_transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.transform(X_test[COL_NUM])
```

```python
#라벨인코딩 결과는 비선형 모델 사용 => 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_tr,y_tr.values.ravel()) ##y_tr.values.ravel() -> y_tr의 값을 1차원 배열로 변환
```

```python
#평가 roc_auc_score
from sklearn.metrics import roc_auc_score

y_pred_rf = model_rf.predict_proba(X_val) #test
#predict_proba 출력 형태 [0,1] => 0번 클래스에 대한 예측 확률, 1번 클래스에 대한 예측 확률 합은 1
```

```python
score_rf = roc_auc_score(y_val,y_pred_rf[:,1]) #y_pred_rf 의 모든 행, 1열 (1 클래스에 대한 예측 확률, 은하 존재 o)
```

```python
score_rf
```

```python
pred = model_rf.predict_proba(X_test)[:,1]
pd.DataFrame({'target':pred}).to_csv('030000.csv',index=False)
```
