##1-5번
Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하시오. 단, 이상치 범위는 (평균 + 2*표준편차) 이상이거나 (평균 - 2*표준편차) 이하이다.

```python
data
```

```python
sl_std=data['sepal_length'].std()
sl_std
```

```python
def outlier(x):
  x_mean = x.mean()
  x_std = x.std()
  return str(len(x[(x>(x_mean+2*x_std))|(x<(x_mean-2*x_std))]))

outlier(data['sepal_length'])
```

##1-5번
Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위수 차를 구하고, class 2에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오

```python
data
```

```python
data['class'].value_counts(dropna=False)
```

```python
data_class_1['class'].value_counts(dropna=False)
```

```python
data_class_2=data[data['class']==2]
data_class_2['class'].value_counts(dropna=False)
```

```python
def iqr(x):
  q1,q3 = x.quantile([0.25,0.75])
  iqr = q3-q1
  return iqr
```

```python
iqr1=iqr(data_class_1['proline'])
```

```python
iqr2=iqr(data_class_2['proline'])
```

```python
abs(iqr1-iqr2)
```
